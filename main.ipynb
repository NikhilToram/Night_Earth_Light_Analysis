### Light Distribution
#### This section shows us histogram distribution of the night light intensity along the decade for Germany and India
##### We see the range in the distribution increases as the years pass by. and the median night emission increases

We may interpret the shift in median as the progress in economic and social growth and the decrease in peak height as increase the energy use inequality

Both the countries seem to be on a similar trend in the light emission distribution curves.

The consistent increase in median light emission also questions if we need regulations on light pollution

### Trends of economic quantiles
#### Here we see the night light trends in the top and bottom quantiles for the countries, the cutoff being 99.5%

Unlike india in case of germany while there is a clear disconnection between the top light emission quantile and the rest. The top economic areas seem to defy the general tredns by a significant margin. This may be because of the different stages of economic growth these both countries are in...
